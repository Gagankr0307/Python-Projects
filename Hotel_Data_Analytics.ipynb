```python
# Check_in_date Format is D/M/YYYY will change to DD-MM-YY
```

```python
def parse_date(date_str):
    for fmt in ('%m/%d/%Y', '%d-%m-%y'):
        try:
            return pd.to_datetime(date_str, format=fmt)
        except ValueError:
            pass
    return pd.to_datetime(date_str)
```

```python
df_bookings_all['check_in_date'] = df_bookings_all['check_in_date'].apply(parse_date)
```

```python
df_bookings_all['check_in_date'] = df_bookings_all['check_in_date'].dt.strftime('%m-%d-%y')
```
